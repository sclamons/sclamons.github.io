```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

import os
```

```python
df = pd.read_csv('200407_cases_by_US_state.csv')
df.head()
```

```python
for state in df.state.unique():
    plt.clf()
    state_df = df[df.state == state]
    sns.lineplot(data = state_df, x = "date", y = "cases")
    plt.title(f"{state} COVID cases")
    plt.xlabel("Date")
    plt.ylabel("# Cases (official)")
    plt.show()
#     plt.savefig(os.path.join("plots",f"{state}_lineplot.png"), dpi = 300)
```
